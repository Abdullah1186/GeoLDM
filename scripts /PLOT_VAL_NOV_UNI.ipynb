```python
import matplotlib.pyplot as plt
import numpy as np
from rdkitfunctions_remake import rdkitfun 
import pandas as pd 



# Example data
datasets = ['QM9', 'OE62']
models = ['Gschnet', 'GeoLDM']
metrics = ['Novelty', 'Uniqueness','Validity' ]

# Example metric values for each model and dataset
data = {
    'QM9': {'Gschnet': [], 'GeoLDM': []},
    'OE62': {'Gschnet':[] }
}



for i in datasets:
    novelty, uniqueness, validity = rdkitfun(i,'Gschnet')
    data[i]['Gschnet']= [novelty,uniqueness,validity]

for i in datasets[:1]:
    novelty, uniqueness, validity = rdkitfun(i,'GeoLDM')
    data[i]['GeoLDM']= [novelty,uniqueness,validity]   

pd.DataFrame(data)



```

```python
# Parameters for the bar chart
x = np.arange(len(metrics))  # x locations for the groups
width = 0.2  # Width of the bars

# Create the plot
fig, ax = plt.subplots(figsize=(20, 10))

# Plot bars for QM9
ax.bar(x , data['QM9']['Gschnet'], width, label='QM9-gschnet')
ax.bar(x + width, data['QM9']['GeoLDM'], width, label='QM9-GeoLDM')

# Plot bars for OE62
ax.bar(x - width, data['OE62']['Gschnet'], width, label='OE62-gschnet', alpha=0.5)
# ax.bar(x + width/2, data['OE62']['GeoLDM'], width, label='OE62-GeoLDM', alpha=0.5)

# Customizing the chart
ax.set_xlabel('Metrics')
ax.set_ylabel('Values')
ax.set_title('Grouped Bar Chart of Metrics for Models and Datasets')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
